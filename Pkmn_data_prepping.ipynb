```python
import pandas as pd
import numpy as py
```

```python
df=pd.ExcelFile('input_pkmn_stats_and_evolutions.xlsx')
df_pks=df.parse(0)
df_pke=df.parse(1)
```

```python
df_pke.shape
```

```python
pd.options.display.max_rows=None
df_pks.head()
```

```python
# delete columns 'height','weight','evolves_from'
df_pks1=df_pks.drop(columns=['height','weight','evolves_from'])
df_pks1.head()
```

```python
# calculate combat_power
df_pks1['combat_power']=df_pks1.iloc[:,3:9].sum(axis=1)
df_pks1
```

```python
df_pks1=df_pks1.drop(columns=['hp','attack','defense','special_defense','speed'])
```

```python
df_pks1=df_pks1.drop(columns='special_attack')
```

```python
df_pks1
```

```python
# If a Pokémon doesn't evolve remove it from the dataset 
index=df_pke.isna().sum(axis=1)<2
index.head()
```

```python
df_pke1=df_pke[index]
df_pke1.shape
```

```python
# join df_pke1 with df_pks1 on Stage 1
df_join=df_pke1.merge(df_pks1, left_on='Stage_1', right_on='name', how='left')
df_join=df_join.drop('name',axis=1)
df_join.rename(columns={'combat_power':'intial_combat_power'}, inplace=True)
df_join.head()
```

```python
df_join['intial_combat_power'].isna().sum()
```

```python
# join df_join with df_pks1 on Stage 3
df_pks1=df_pks1.drop(columns=['pokedex_number','gen_introduced'])
df_join1=df_join.merge(df_pks1, left_on='Stage_3', right_on='name', how='left')
df_join1=df_join1.drop('name',axis=1)
df_join1.rename(columns={'combat_power':'final_combat_power'}, inplace=True)
```

```python
# join df_join with df_pks1 on Stage 2
df_join2=df_join1.merge(df_pks1, left_on='Stage_2', right_on='name', how='left')
df_join2=df_join2.drop('name',axis=1)
df_join2.rename(columns={'combat_power':'stage2_combat_power'}, inplace=True)
```

```python
df_join2
```

```python
# find final_combat_power==nan
index1=df_join2['final_combat_power'].isna()
df_join2.loc[index1,'final_combat_power']=df_join2.loc[index1,'stage2_combat_power']
#df_join2['final_combat_power'].isna().sum()
```

```python
# change data type of final_combat_power
df_join2['final_combat_power']=df_join2['final_combat_power'].astype(int)
```

```python
# drop stage2_combat_power
df_join2=df_join2.drop('stage2_combat_power',axis=1)
```

```python
df_join2['combat_power_increase']=df_join2['final_combat_power']/df_join2['intial_combat_power']-1
df_join2.head()
```

```python
# Sort the dataset, ascending by percentage increase
df_join2.sort_values(by='combat_power_increase',inplace=True)
df_join2
```

```python
# output data
df_join2.to_excel('output_data_Jupyter.xlsx', index=False)
```
